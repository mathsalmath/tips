#What is the total revenue by day?


```python
tips.groupby('day')['total_bill'].sum()
```

```python
fig = px.histogram(tips, x='day', y='total_bill', color='day', template='plotly_white', text_auto=True, title='Total Revenue')
fig.update_layout(height = 750)
fig.show()
```

#What was the revenue generated from lunch and dinner?

```python
tips.groupby('time')['total_bill'].sum()
```

```python
fig = px.histogram(tips, x='time', y='total_bill', color='time', template='plotly_white', text_auto=True, title='Total Revenue Lunch/Dinner')
fig.update_layout(height = 750)
fig.show()
```

#What was the revenue generated by each gender (m/f)?

```python
tips.groupby('sex')['total_bill'].sum()
```

```python
fig = px.histogram(tips, x='sex', y='total_bill', color='sex', template='plotly_white', text_auto=True, title='Total Revenue by Gender')
fig.update_layout(height = 750)
fig.show()
```

#What was the total revenue generated by each gender for lunch/dinner?

```python
pd.pivot_table(tips, index='sex', columns='time', values='total_bill', aggfunc='sum')
```

```python
fig = px.histogram(tips, x='sex', y='total_bill', color='time', barmode='group', template='plotly_white',
             title='Total Revenue by Gender Lunch/Dinner', text_auto=True)
fig.update_layout(height = 650)
fig.show()
```

#Number of guests (party size) analysis

##Most popular and least popular party(guest) size

*   most popular (party of 2)
*   Least popular (least popular party of 1)



```python
tips.groupby('size')['total_bill'].sum().sort_values(ascending=False)
```

```python
fig = px.histogram(tips, x='size', color='size', template='plotly_white', text_auto=True, title='Popular Party Size')
fig.update_layout(height = 750)
fig.show()
```

#How many male/female smokers visited the resturant?

```python
tips.groupby('sex')['smoker'].value_counts()
```

```python
fig = px.histogram(tips, x='sex', color='smoker', template='plotly_white', text_auto=True, title = 'Smoker Count')
fig.update_layout(height = 750)
fig.show()
```

#sunburst chart

```python
px.sunburst(tips, path=['sex', 'smoker', 'day', 'time'], values='total_bill', title='Sunburst Chart')
```
